```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
pkg = mp.jupyter.open_package()
#pkg = mp.jupyter.open_source_package()
pkg
```

```python
c20m = pkg.resource('county_20m').geoframe()
s20m = pkg.resource('state_20m').geoframe()
```

```python
fig, ax = plt.subplots(figsize=(12,8))  
c20m[~c20m.stusab.isin(['HI','AK', 'PR'])].plot(ax=ax, facecolor="cornflowerblue", edgecolor='royalblue')
s20m[~s20m.stusab.isin(['HI','AK', 'PR'])].plot(ax=ax, facecolor="none", edgecolor='black', lw=0.7)
plt.savefig('continental.png')
```
